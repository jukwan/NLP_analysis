```python
path= 'F:\\Dropbox\\2018\\WordClouds' #for load the data
#path ='C:\\Users\\user\\Dropbox\\Flight\\data\\cleaned_data'

df= pd.read_csv(path+'\\Book1.csv')
```

## data to list

```python
data=df.values.T.ravel().tolist()

```

## list to string

```python
string = ' '.join(str(x) for x in data)
```

### remove un necessary word by picking nouns

```python
from konlpy.tag import Twitter; t = Twitter()
```

```python
token2= t.nouns(string)
```

```python
len(token2)
```

#### Double check the wierd meaningless words

```python
stop_words = ['기','용','구', '제','고','전','빈','밀']
```

remove meaningless words in above list

```python
token2 = [ each_word for each_word in token2 if each_word not in stop_words]
```

### Tokenizing the filterd nouns
and keep the most frequent one (ranks)
and Dictionize {nouns: freq}

```python
ko = nltk.Text(token2)
clouds= ko.vocab().most_common(500)
temps = dict(clouds)
```

```python
ko.plot(30)
```

### Plot in the graph using word clouds.

```python
import matplotlib.pyplot as plt
import platform

```

```python
worldcould = WordCloud(font_path = 'c:/Windows/Fonts/malgun.ttf',relative_scaling = 0.2, 
                       background_color= 'white').generate_from_frequencies(temps)
```

```python
plt.figure(figsize=(50,50))
plt.imshow(worldcould, interpolation ='bilinear')
plt.axis("off")
plt.show()
```
